```python
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Load dan proses data
data = pd.read_csv('Dataset duduk berdiri.csv')
X = data[['ax', 'ay', 'az']]
y = data['Gerakan']

# Buat mapping class
classes = list(set(y))
mapping = dict(zip(classes, range(len(classes))))
y = y.replace(mapping)

# Buat tensor
X = tf.convert_to_tensor(X)
y = tf.convert_to_tensor(y)
y = tf.one_hot(y, depth=len(classes))

# Buat Dataset
dataset = tf.data.Dataset.from_tensor_slices((X, y))

# Split menjadi 60% latih dan 40% uji
batch_size = 32
train_size = int(len(dataset) * 0.6)
test_size = len(dataset) - train_size
train_dataset = dataset.skip(test_size).take(train_size).batch(batch_size)
test_dataset = dataset.skip(train_size).batch(batch_size)

# Model
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(10, input_shape=(3,)))
model.add(tf.keras.layers.Dense(10, activation='relu'))
model.add(tf.keras.layers.Dense(len(classes), activation='softmax'))

# Compile dan latih model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(train_dataset, epochs=100)

# Evaluasi
results = model.evaluate(test_dataset)
print("Test accuracy:", results[1])

for x, y in test_dataset:
  y_true = y

y_pred = model.predict(test_dataset)
y_pred_classes = np.argmax(y_pred, axis=1)

accuracy = accuracy_score(y_true, y_pred_classes)
cm = confusion_matrix(y_true, y_pred_classes)

print('Accuracy:', accuracy)
print(cm)
```
